```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import pandas as pd

# Генерация данных
np.random.seed(42)
x = np.linspace(0, 1, 200)
y_linear = x*3 + 2
y_linear = np.exp(y_linear)
y = y_linear + (np.random.randint(0,10,200))

# Сохранение данных в CSV
data = pd.DataFrame({'x': x, 'y': y})
data.to_csv('generated_data_new.csv', index=False)

# Визуализация данных
plt.scatter(x, y, color='blue', label='Generated Data')
plt.plot(x, y_linear, color='red', label='Original Linear Function')
plt.legend()
plt.title('Generated Data with Noise')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures

# Преобразование данных
x = x.reshape(-1, 1)
poly = PolynomialFeatures(degree=2)
x_poly = poly.fit_transform(x)

# Обучение модели линейной регрессии
regressor = LinearRegression()
regressor.fit(x_poly, y)
y_pred = regressor.predict(x_poly)

# Визуализация результатов
plt.scatter(x, y, color='blue', label='Generated Data')
plt.plot(x, y_pred, color='green', label='Linear Regression with Polynomial Features')
plt.legend()
plt.title('Linear Regression with Polynomial Features')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()

# Вычисление MSE
mse = mean_squared_error(y, y_pred)
mse
```
