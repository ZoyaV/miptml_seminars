## Почему полезно логировать эксперименты?

Логирование экспериментов в машинном обучении необходимо для обеспечения прозрачности и воспроизводимости результатов. Сохраняя информацию об окружении, коде и результатах экспериментов, вы можете легко воспроизвести результаты и сравнить различные запуски. Это особенно важно при работе в команде, так как это позволяет лучше сотрудничать и документировать результаты. Кроме того, логирование экспериментов помогает визуализировать и понимать производительность моделей в течение времени, что позволяет выявлять тенденции, принимать обоснованные решения о том, какие модели следует использовать, и оптимизировать модели для лучшей производительности.

## Что можно делать с помощью wandb?

W&B (Weights & Biases) - это платформа для мониторинга и анализа экспериментов в области машинного обучения. Она предоставляет возможность визуализировать результаты в режиме реального времени и хранить их в облаке для дальнейшего доступа


С помощью W&B вы можете:

* Визуализировать и анализировать результаты экспериментов в режиме реального времени.

* Сохранять и делиться результатами экспериментов в облаке.

* Отслеживать переменные, метрики и снимки весов моделей.

* Повторять и воспроизводить эксперименты с легкостью.

* Работать с командой в режиме реального времени, делясь результатами и обсуждая их.

Каждый раз, когда мы вызываем fit мы обучаем модель на всех данных, это аналог одной эпохи. Мы логируем accuracy после каждой эпохи, чтобы мы могли наблюдать за процессом обучения в WandB. Мы также логируем номер текущей эпохи, чтобы мы могли увидеть, как меняется accuracy со временем.

## Пример

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torchvision import datasets, transforms
import wandb
import copy


# Инициализация wandb
wandb.init(project="image_classification")

# Конфигурация
config = wandb.config
config.learning_rate = 0.001
config.epochs = 10
config.batch_size = 64

# Загрузка и предобработка данных
transform = transforms.Compose([transforms.ToTensor(),
                                transforms.Normalize((0.5,), (0.5,))])

trainset = datasets.CIFAR10(root='./data', train=True,
                            download=True, transform=transform)
trainloader = torch.utils.data.DataLoader(trainset, batch_size=config.batch_size,
                                          shuffle=True)

testset = datasets.CIFAR10(root='./data', train=False,
                           download=True, transform=transform)
testloader = torch.utils.data.DataLoader(testset, batch_size=config.batch_size,
                                         shuffle=False)

# Определение модели CNN
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.conv1 = nn.Conv2d(3, 6, 5)
        self.pool = nn.MaxPool2d(2, 2)
        self.conv2 = nn.Conv2d(6, 16, 5)
        self.fc1 = nn.Linear(16 * 5 * 5, 120)
        self.fc2 = nn.Linear(120, 84)
        self.fc3 = nn.Linear(84, 10)

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = x.view(-1, 16 * 5 * 5)
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x

net = Net()

# Определение функции потерь и оптимизатора
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(net.parameters(), lr=config.learning_rate, momentum=0.9)

# Функция для обучения и валидации модели
def train_model(epochs, trainloader, testloader, criterion, optimizer, model, config):
    best_accuracy = 0.0
    for epoch in range(epochs):
        running_loss = 0.0
        for i, data in enumerate(trainloader, 0):
            inputs, labels = data
            optimizer.zero_grad()

            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            running_loss += loss.item()
            if i % 1000 == 999:
                print(f"[{epoch + 1}, {i + 1}] loss: {running_loss / 1000:.3f}")
                running_loss = 0.0

        # Валидация
        correct = 0
        total = 0
        with torch.no_grad():
            for data in testloader:
                images, labels = data
                outputs = model(images)
                _, predicted = torch.max(outputs.data, 1)
                total += labels.size(0)
                correct += (predicted == labels).sum().item()

        accuracy = 100 * correct / total
        print(f"Accuracy of the network on the 10000 test images: {accuracy}%")

        # Логирование в wandb
        wandb.log({"epoch": epoch, "loss": running_loss, "accuracy": accuracy})

        # Сохранение лучшей модели
        if accuracy > best_accuracy:
            best_accuracy = accuracy
            best_model_wts = copy.deepcopy(model.state_dict())
            torch.save(best_model_wts, "best_model.pth")
            wandb.save("best_model.pth")

    print('Finished Training')

# Обучение модели
train_model(config.epochs, trainloader, testloader, criterion, optimizer, net, config)

# Загрузка лучшей модели и создание артефакта в wandb
artifact = wandb.Artifact('model', type='model')
artifact.add_file('best_model.pth')
wandb.log_artifact(artifact)
```

### Пример Sweep

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torchvision import datasets, transforms
import wandb
import copy

# Определение модели CNN
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.conv1 = nn.Conv2d(3, 6, 5)
        self.pool = nn.MaxPool2d(2, 2)
        self.conv2 = nn.Conv2d(6, 16, 5)
        self.fc1 = nn.Linear(16 * 5 * 5, wandb.config.hidden_size)
        self.fc2 = nn.Linear(wandb.config.hidden_size, 84)
        self.fc3 = nn.Linear(84, 10)

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = x.view(-1, 16 * 5 * 5)
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x

# Функция для обучения модели
def train():
    # Инициализация wandb
    wandb.init()

    # Загрузка и предобработка данных
    transform = transforms.Compose([transforms.ToTensor(),
                                    transforms.Normalize((0.5,), (0.5,))])

    trainset = datasets.CIFAR10(root='./data', train=True,
                                download=True, transform=transform)
    trainloader = torch.utils.data.DataLoader(trainset, batch_size=wandb.config.batch_size,
                                              shuffle=True)

    testset = datasets.CIFAR10(root='./data', train=False,
                               download=True, transform=transform)
    testloader = torch.utils.data.DataLoader(testset, batch_size=wandb.config.batch_size,
                                             shuffle=False)

    net = Net()

    # Определение функции потерь и оптимизатора
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.SGD(net.parameters(), lr=wandb.config.learning_rate, momentum=0.9)

    # Обучение и валидация модели
    best_accuracy = 0.0
    for epoch in range(wandb.config.epochs):
        running_loss = 0.0
        for i, data in enumerate(trainloader, 0):
            inputs, labels = data
            optimizer.zero_grad()

            outputs = net(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            running_loss += loss.item()

        # Валидация
        correct = 0
        total = 0
        with torch.no_grad():
            for data in testloader:
                images, labels = data
                outputs = net(images)
                _, predicted = torch.max(outputs.data, 1)
                total += labels.size(0)
                correct += (predicted == labels).sum().item()

        accuracy = 100 * correct / total
        print(f"Epoch {epoch+1}: Accuracy on test data: {accuracy}%")

        # Логирование в wandb
        wandb.log({"epoch": epoch, "loss": running_loss / len(trainloader), "accuracy": accuracy})

        # Сохранение лучшей модели
        if accuracy > best_accuracy:
            best_accuracy = accuracy
            best_model_wts = copy.deepcopy(net.state_dict())
            torch.save(best_model_wts, "best_model.pth")
            wandb.save("best_model.pth")

    print('Finished Training')

    # Загрузка лучшей модели и создание артефакта в wandb
    artifact = wandb.Artifact('model', type='model')
    artifact.add_file('best_model.pth')
    wandb.log_artifact(artifact)

# Настройка и запуск Sweeps
def sweep_train():
    sweep_config = {
        'method': 'random',
        'metric': {
            'name': 'accuracy',
            'goal': 'maximize'
        },
        'parameters': {
            'learning_rate': {
                'values': [0.001, 0.0005, 0.0001]
            },
            'batch_size': {
                'values': [32, 64]
            },
            'hidden_size': {
                'values': [120, 240]
            },
            'epochs': {
                'value': 5
            }
        }
    }

    sweep_id = wandb.sweep(sweep_config, project="image_classification")
    wandb.agent(sweep_id, train)

# Запуск Sweeps
sweep_train()
```
