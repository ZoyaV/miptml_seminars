## Shap для анализа табличных данных

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, accuracy_score
from sklearn.utils import class_weight
import numpy as np
```

### Данные задачи

https://www.kaggle.com/datasets/rashikrahmanpritom/heart-attack-analysis-prediction-dataset



Целью этой задачи является разработка модели машинного обучения, которая может предсказывать риск сердечного приступа у пациентов на основе различных медицинских показателей. Датасет включает в себя данные такие как уровень холестерина, возраст, пол, тип боли в груди, максимальный пульс и другие важные медицинские показатели. Анализируя эти данные, модель должна быть способна определить вероятность наступления сердечного приступа у конкретного индивидуума.

```python
import pandas as pd

# Load the dataset
file_path = 'heart.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataframe
data.head()
```

### Описание признаков
* age: Возраст пациента
* sex: Пол пациента (1 = мужчина, 0 = женщина)
* cp: Тип боли в груди (значения от 1 до 4)
* Значение 1: типичная стенокардия
 * Значение 2: атипичная стенокардия
 * Значение 3: нестенокардическая боль
 * Значение 4: бессимптомная
* trtbps: Артериальное давление в покое (в мм рт. ст.)
* chol: Уровень холестерина в мг/дл
* fbs: Уровень сахара в крови натощак > 120 мг/дл (1 = да; 0 = нет)
* restecg: Результаты электрокардиографии в покое (значения 0, 1, 2)
* Значение 0: нормальные
* Значение 1: наличие аномалии волны ST-T (инверсии волны T и/или подъём или депрессия ST > 0.05 мВ)
* Значение 2: вероятность или определённая гипертрофия левого желудочка по критериям Эстеса
* thalachh: Максимальный достигнутый пульс
* exng: Стенокардия, вызванная физической нагрузкой (1 = да; 0 = нет)
* oldpeak: Депрессия ST, вызванная упражнениями по сравнению с отдыхом
* slp: Наклон пика сегмента ST при пиковой нагрузке
* caa: Количество крупных сосудов (0-3), окрашенных при флюороскопии
* thall: Результаты таллиевого стресс-теста (значения от 1 до 3)
* output: Целевая переменная, где 1 указывает на больший шанс сердечного приступа

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Check for any missing values in the dataset
missing_values = data.isnull().sum()

# Summary statistics for the dataset
summary_statistics = data.describe()

# Checking the balance of the target variable
class_distribution = data['output'].value_counts(normalize=True)

# Visualizing the distribution of the target variable
plt.figure(figsize=(8, 4))
sns.countplot(x='output', data=data)
plt.title('Distribution of the Target Variable (Heart Attack Chance)')
plt.show()

(missing_values, summary_statistics, class_distribution)
```

### Обучение модели XGBClassifier

```python
# Separate the features and the target variable
X = data.drop('output', axis=1)
y = data['output']

# Scaling the features - though it's not always necessary for tree-based models,
# it can be beneficial for gradient boosting models to have features on the same scale
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Compute the class weights
class_weights = class_weight.compute_class_weight('balanced', classes=np.unique(y), y=y)
class_weights = dict(enumerate(class_weights))

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42, stratify=y)

```

```python
# Adjust the model parameters to further speed up training
# We increase the learning rate to make the model converge faster
xgb_model_faster = XGBClassifier(n_estimators=10, max_depth=2,
                                 learning_rate=0.5, scale_pos_weight=class_weights[1],
                                 use_label_encoder=False,
                                 eval_metric='logloss', random_state=42)

# Train the model with the faster parameters
xgb_model_faster.fit(X_train, y_train)

# Make predictions on the test set with the faster model
y_pred_faster = xgb_model_faster.predict(X_test)

# Evaluate the faster model
accuracy_faster = accuracy_score(y_test, y_pred_faster)
class_report_faster = classification_report(y_test, y_pred_faster)
```

```python
print(class_report_faster)
```

### Выяление важности признаков

```python
import shap
# Create the SHAP explainer using the trained model
explainer = shap.Explainer(xgb_model_faster)
# Calculate SHAP values for the test set
shap_values = explainer(X_test)

# Summary plot of SHAP values for all features
shap.summary_plot(shap_values, X_test, feature_names=data.columns[:-1])
```

#### Интерпретация графика

Признаки, расположенные выше на графике, имеют большее влияние на модель. *Например, cp (тип боли в груди) является наиболее влиятельным признаком.*

Цвет точек указывает на значение признака для данного наблюдения: высокие значения признака обозначены красным цветом, низкие значения — синим.

Положение точек по горизонтали показывает влияние признака на прогнозную вероятность. Признаки, для которых точки смещены вправо (положительные значения SHAP), увеличивают вероятность исхода модели (например, вероятность сердечного приступа), в то время как признаки с точками, смещенными влево (отрицательные значения SHAP), уменьшают её.

Горизонтальное распределение точек для каждого признака показывает степень его влияния на различные наблюдения. Широкое распределение означает, что признак оказывает различное влияние на разные наблюдения.

#### Интерпретация

- **Тип боли в груди (`cp`)**: Высокие значения этого признака (красные точки, смещенные вправо) увеличивают вероятность положительного исхода прогноза (в данном контексте — риск сердечного приступа).

- **Результаты таллиевого стресс-теста (`thall`)**: Низкие значения этого признака (синие точки, смещенные вправо) увеличивают риск сердечного приступа, в то время как высокие значения (красные точки, смещенные влево) уменьшают его.

- **Количество крупных сосудов (`caa`)**: Подобно `thall`, низкие значения `caa` увеличивают риск, а высокие значения уменьшают риск сердечного приступа.

- **Депрессия ST (`oldpeak`)**: Высокие значения этого признака (красные точки) имеют тенденцию увеличивать риск сердечного приступа.

- **Наклон пика сегмента ST при пиковой нагрузке (`slp`)**: Изменения этого признака могут как увеличивать, так и уменьшать риск сердечного приступа, что отражено в смешанном распределении точек.

- **Уровень холестерина (`chol`)**: Имеет смешанное влияние на прогнозы, с некоторыми высокими значениями увеличивающими и некоторыми уменьшающими риск.

- **Максимальный достигнутый пульс (`thalachh`)**: Низкие значения этого признака имеют тенденцию увеличивать риск сердечного приступа.

- **Пол (`sex`)**: В этом случае также наблюдается смешанное влияние на прогнозы, что может указывать на то, что риск сердечного приступа нелинейно связан с полом пациента.

- **Стенокардия, вызванная физической нагрузкой (`exng`)**: Высокие значения этого признака (красные точки, смещенные влево) имеют тенденцию уменьшать риск сердечного приступа.

- **Возраст (`age`)** и **артериальное давление в покое (`trtbps`)**: Имеют смешанное влияние на прогнозы.

- **Уровень сахара в крови натощак (`fbs`)**: Показывает незначительное влияние на прогнозы, с некоторым смещением в сторону увеличения риска при высоких значениях.

- **Результаты электрокардиографии в покое (`restecg`)**: Не оказывают значительного влияния на прогнозы.



### Интепретация для конкретного предсказания

```python
shap.initjs()

# Получение SHAP значений для конкретного примера
shap_values_single = explainer(X_test[0:1, :])

# Визуализация SHAP значений с помощью force plot
# В новых версиях SHAP вам нужно передать объект Explanation напрямую
shap.force_plot(shap_values_single.base_values[0], shap_values_single.values[0], X_test[0:1, :], feature_names=X.columns.tolist())
```

На представленном SHAP force plot графике показан вклад каждого признака в конкретное предсказание модели. График состоит из стрелок, которые указывают в сторону, повышающую или понижающую предсказание модели:

- **Синие стрелки** представляют признаки, которые влияют на уменьшение предсказанного значения (или уменьшают вероятность положительного класса в задаче классификации).
- **Красные стрелки** представляют признаки, которые увеличивают предсказанное значение (или повышают вероятность положительного класса).

Каждая стрелка имеет размер, соответствующий величине вклада признака. Базовое значение (base value) — это значение, которое было бы предсказано моделью, если бы ни один из признаков не был известен. Оттуда каждая стрелка изменяет предсказание от базового значения в сторону увеличения или уменьшения. Итоговое предсказание (f(x)) — это сумма базового значения и всех этих вкладов.



На основе SHAP force plot, мы можем интерпретировать вклад каждого признака в прогнозируемый исход следующим образом:

- **thal**: Наиболее влиятельный признак, который сильно уменьшает вероятность положительного исхода (в контексте медицинских данных, это может означать снижение вероятности наличия болезни).
- **oldpeak**: Также уменьшает вероятность положительного исхода, но его влияние меньше, чем у `thal`.
- **caa**: Имеет небольшое положительное влияние на вероятность исхода, что может указывать на увеличение риска с увеличением значения этого признака.
- **cp**: Слегка уменьшает вероятность положительного исхода.
- **thalachh**: Имеет отрицательное влияние на прогноз, уменьшая вероятность положительного исхода.
- **chol**: Минимально влияет на увеличение вероятности положительного исхода.
- **exng**: Увеличивает вероятность положительного исхода, хотя его влияние несколько меньше, чем у `chol`.
- **slp**: Уменьшает вероятность положительного исхода.
- **restecg**: Также уменьшает вероятность положительного исхода, но его влияние меньше, чем у `slp`.

Базовое значение (base value) представляет собой среднее предсказание модели по всему набору данных, а итоговое предсказание (f(x)) является суммой базового значения и вклада от всех признаков для данного примера. В данном случае итоговое предсказание смещено в сторону уменьшения вероятности положительного исхода из-за значительного вклада признака `thal`.

### Улучшение модели

```python
from sklearn.model_selection import GridSearchCV

# Define a parameter grid to search for the best parameters
param_grid = {
    'max_depth': [3, 4, 5],
    'learning_rate': [0.1, 0.01, 0.05],
    'n_estimators': [50, 100, 200],
    'subsample': [0.8, 0.9, 1.0],
    'colsample_bytree': [0.8, 0.9, 1.0],
}

# Initialize the XGBClassifier
xgb_model = XGBClassifier(use_label_encoder=False, eval_metric='logloss')

# Create the GridSearchCV object
grid_search = GridSearchCV(estimator=xgb_model, param_grid=param_grid, cv=3, n_jobs=-1, verbose=2, scoring='accuracy')

# Fit the grid search to the data
grid_search.fit(X_train, y_train)

# Get the best parameters
best_parameters = grid_search.best_params_

# Train the final model with the best parameters found
final_model = XGBClassifier(**best_parameters)
final_model.fit(X_train, y_train)

# Output the best parameters and return the trained final model
(best_parameters, final_model)
```

```python
# Make predictions on the test set with the faster model
y_pred_fin = final_model.predict(X_test)

# Evaluate the faster model
accuracy_fin = accuracy_score(y_test, y_pred_fin)
class_report_fin = classification_report(y_test, y_pred_fin)
```

## SHAP для текстового анализа

```python
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer
import torch

model = AutoModelForSeq2SeqLM.from_pretrained("google/flan-t5-base")
tokenizer = AutoTokenizer.from_pretrained("google/flan-t5-base")
```

```python
sentence = "You're a good guy"
inputs = tokenizer(f"Sentiment to the sentence {sentence} positive/negative/neutral:", return_tensors="pt")
outputs = model.generate(**inputs)
print(tokenizer.batch_decode(outputs, skip_special_tokens=True))
```

```python
import numpy as np

def predict_probabilities(texts):
    print(texts)
    # texts - это список предложений
    predictions = []
    for text in texts:
        # Токенизация и подача текста в модель
        inputs = tokenizer(f"Sentiment to the sentence {text} positive/negative/neutral:", return_tensors="pt")
        outputs = model.generate(**inputs)
        # Декодирование результата
        decoded_output = tokenizer.decode(outputs[0], skip_special_tokens=True)
        # Преобразование декодированных меток в вероятности
        if 'positive' in decoded_output:
            predictions.append(1)
        elif 'negative' in decoded_output:
            predictions.append(-1)
        elif 'neutral' in decoded_output:
            predictions.append(0)
        else:
            predictions.append(0)  # Например, если метка не определена
    return np.array(predictions)

# Используем Text masker для текстовых данных
masker = shap.maskers.Text(tokenizer)

# Создание экземпляра SHAP Explainer
explainer = shap.Explainer(predict_probabilities, masker)
```

```python
# Пример использования
texts = ["You are a good guy"]
shap_values = explainer(texts)
shap.initjs()
shap.plots.text(shap_values[:, :])
```

### SHAP для классификации изображений

```python
import json

import numpy as np
import torch
import torchvision

import shap
```

```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
```

```python
model = torchvision.models.mobilenet_v2(pretrained=True, progress=False)
model.to(device)
model.eval()
X, y = shap.datasets.imagenet50()
```

```python

# Getting ImageNet 1000 class names
url = "https://s3.amazonaws.com/deep-learning-models/image-models/imagenet_class_index.json"
with open(shap.datasets.cache(url)) as file:
    class_names = [v[1] for v in json.load(file).values()]
print("Number of ImageNet classes:", len(class_names))
# print("Class names:", class_names)
```

```python
# Prepare data transformation pipeline

mean = [0.485, 0.456, 0.406]
std = [0.229, 0.224, 0.225]


def nhwc_to_nchw(x: torch.Tensor) -> torch.Tensor:
    if x.dim() == 4:
        x = x if x.shape[1] == 3 else x.permute(0, 3, 1, 2)
    elif x.dim() == 3:
        x = x if x.shape[0] == 3 else x.permute(2, 0, 1)
    return x


def nchw_to_nhwc(x: torch.Tensor) -> torch.Tensor:
    if x.dim() == 4:
        x = x if x.shape[3] == 3 else x.permute(0, 2, 3, 1)
    elif x.dim() == 3:
        x = x if x.shape[2] == 3 else x.permute(1, 2, 0)
    return x


transform = [
    torchvision.transforms.Lambda(nhwc_to_nchw),
    torchvision.transforms.Lambda(lambda x: x * (1 / 255)),
    torchvision.transforms.Normalize(mean=mean, std=std),
    torchvision.transforms.Lambda(nchw_to_nhwc),
]

inv_transform = [
    torchvision.transforms.Lambda(nhwc_to_nchw),
    torchvision.transforms.Normalize(
        mean=(-1 * np.array(mean) / np.array(std)).tolist(),
        std=(1 / np.array(std)).tolist(),
    ),
    torchvision.transforms.Lambda(nchw_to_nhwc),
]

transform = torchvision.transforms.Compose(transform)
inv_transform = torchvision.transforms.Compose(inv_transform)
```

```python
def predict(img: np.ndarray) -> torch.Tensor:
    img = nhwc_to_nchw(torch.Tensor(img))
    img = img.to(device)
    output = model(img)
    return output
```

```python
# Check that transformations work correctly
Xtr = transform(torch.Tensor(X))
out = predict(Xtr[1:3])
classes = torch.argmax(out, axis=1).cpu().numpy()
print(f"Classes: {classes}: {np.array(class_names)[classes]}")
```

```python
topk = 4
batch_size = 50
n_evals = 1000

# define a masker that is used to mask out partitions of the input image.
masker_blur = shap.maskers.Image("blur(128,128)", Xtr[0].shape)

# create an explainer with model and image masker
explainer = shap.Explainer(predict, masker_blur, output_names=class_names)

# feed only one image
# here we explain two images using 100 evaluations of the underlying model to estimate the SHAP values
shap_values = explainer(
    Xtr[1:2],
    max_evals=n_evals,
    batch_size=batch_size,
    outputs=shap.Explanation.argsort.flip[:topk],
)
```
